Импорт библиотек

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import parallel_coordinates
from math import pi
import squarify
```

Загрузка датасета

```python
df = pd.read_csv("winequalityN.csv")
print(df)
```

DENSITY

```python
sns.kdeplot(x=df['quality'],y=df['pH'])
```

Boxplot

```python
sns.boxplot(x=df['quality'],y=df['fixed acidity'])
```

Density 2D

```python
sns.kdeplot(x=df['quality'],y=df['alcohol'], cmap="Reds", shade=True, bw_adjust=1)
```

Treemap

```python
numtype = []
white = 0
red = 0
for i in df['type']:
    if i == 'white':
        white += 1
    else:
        red += 1
numtype.append(white)
numtype.append(red)
squarify.plot(sizes=numtype, label=['white','red'], alpha=.8 )
plt.axis('off')
plt.show()
```

Quick start

```python
height = df['quality'].value_counts().to_frame()
bars = df['quality'].unique()
y_pos = np.arange(len(bars))
plt.bar(y_pos, height['quality'])
plt.xticks(y_pos, bars)
plt.show()
```
